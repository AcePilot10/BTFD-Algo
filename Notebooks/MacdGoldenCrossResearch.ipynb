```python
import pandas as pandas
import matplotlib.pyplot as plt
import yfinance
import datetime
import mplfinance as mpf
```

```python
start = datetime.datetime(2016, 1, 1)
end = datetime.datetime(2021, 1, 1)
```

```python
ivv = yfinance.Ticker("IVV")
iyy = yfinance.Ticker("IYY")
ijr = yfinance.Ticker("IJR")
spy = yfinance.Ticker("SPY")
```

```python
ivv_hist = ivv.history(start=start, end=end)
iyy_hist = iyy.history(start=start, end=end)
ijr_hist = ijr.history(start=start, end=end)
spy_hist = spy.history(start=start, end=end)
```

```python
mpf.plot(spy_hist, type='line', mav=(150), title='SPY')
mpf.plot(ivv_hist, type='line', mav=(150), title='IVV')
mpf.plot(iyy_hist, type='line', mav=(150), title='IYY')
mpf.plot(ijr_hist, type='line', mav=(150), title='IJR')
```
